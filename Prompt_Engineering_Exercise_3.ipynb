**Iteration 1: Prompt - You are an online customer service agent at Walmart. Create a Python 3 program that is meant to be run in Colab that creates an email response to a customer requesting an update on a delayed shipment item, specifically a Nintendo Switch console.**



**This is the initial output:**


```python
# Walmart Customer Service Email Generator
# Designed to run in Google Colab (Python 3)

from datetime import datetime, timedelta

def generate_email(customer_name, order_number, original_delivery_date, delay_days=3):
    """
    Generates a professional email response regarding a delayed Nintendo Switch shipment.
    """

    # Calculate new estimated delivery date
    original_date_obj = datetime.strptime(original_delivery_date, "%Y-%m-%d")
    new_delivery_date = original_date_obj + timedelta(days=delay_days)

    email = f"""
Subject: Update on Your Nintendo Switch Order #{order_number}

Dear {customer_name},

Thank you for contacting Walmart Customer Support regarding your Nintendo Switch console order.

We sincerely apologize for the delay in your shipment. Due to unexpected carrier delays, your package has not arrived by the originally scheduled delivery date of {original_delivery_date}.

The updated estimated delivery date is {new_delivery_date.strftime('%Y-%m-%d')}. Our logistics team is actively monitoring the shipment to ensure it reaches you as soon as possible.

We understand how exciting it is to receive your Nintendo Switch, and we truly appreciate your patience. If the order does not arrive by the new estimated date or if you have any further concerns, please don't hesitate to contact us.

As a token of our apology, we would be happy to review available compensation options if the delay continues.

Thank you for choosing Walmart. We value your business.

Sincerely,
Walmart Customer Care Team
"""

    return email


# Example usage (you can change these values)
customer_name = "John Smith"
order_number = "WM123456789"
original_delivery_date = "2026-02-23"

# Generate and print the email
email_response = generate_email(customer_name, order_number, original_delivery_date)
print(email_response)
```

**Final Revised Iteration: Prompt - Modify the Python program. Is the response you provided to the customer under 150 words? Does it accurately address the delay? Is it clear and easy to understand? Does it exemplify empathy and professionalism? If not, revise it.**


**This is the revised AI output, which follows the constraints stated:**

```python
# Walmart Customer Service Email Generator (Under 150 Words Version)
# Designed for Google Colab - Python 3

from datetime import datetime, timedelta

def generate_email(customer_name, order_number, original_delivery_date, delay_days=3):
    """
    Generates a professional, empathetic email response
    under 150 words regarding a delayed Nintendo Switch shipment.
    """

    original_date_obj = datetime.strptime(original_delivery_date, "%Y-%m-%d")
    new_delivery_date = original_date_obj + timedelta(days=delay_days)

    email = f"""
Subject: Update on Your Nintendo Switch Order #{order_number}

Dear {customer_name},

Thank you for contacting Walmart regarding your Nintendo Switch order. I sincerely apologize for the delay in your shipment. We understand how important it is to receive your item on time.

Your original delivery date was {original_delivery_date}. Due to an unexpected carrier delay, your updated estimated delivery date is {new_delivery_date.strftime('%Y-%m-%d')}.

Our team is actively tracking your shipment to ensure it arrives as quickly as possible. If your order does not arrive by the new estimated date, please contact us so we can assist further.

We truly appreciate your patience and thank you for choosing Walmart.

Sincerely,
Walmart Customer Care Team
"""

    return email.strip()


def check_word_count(text):
    return len(text.split())


# Example usage
customer_name = "John Smith"
order_number = "WM123456789"
original_delivery_date = "2026-02-23"

email_response = generate_email(customer_name, order_number, original_delivery_date)
word_count = check_word_count(email_response)

print(email_response)
print("\nWord Count:", word_count)

if word_count <= 150:
    print("Status: Email is under 150 words.")
else:
    print("Status: Email exceeds 150 words.")
```
